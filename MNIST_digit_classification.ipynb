```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

in this we will have 10 output nodes and jiski probability highest hogi vhi digit hoga like har note kisi digit k hone ki prob btaega and jis node k digit hone ki rob max chi digit final ans

```python
(X_train,y_train),(X_test,y_test)= keras.datasets.mnist.load_data()
```

In data values should be in similar range that makes aur nn better and converges faster so we will divide all valueswith maxm value

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

1st we have to convert our 2d pixels to 1d arrays ie from 28 x 28 to 784 usingb flatten layer also we use softmax when output is many

also once a increased another hidden layer with 32 nodes my data became overfitted and loss fluctuated

```python
model= Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

in catag... we have to one hot encode our y label and in sparse_catag.... we dont have to

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

```python
model.predict(X_test)
```

uper array me har digit k hone ki probability hai like 1 k hone ki 2 .... since finally we want ki output 1,2 or 9 hai for that we have to get ki array me kis index pe probability highest AAI Hai

```python
y_prob=model.predict(X_test)
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
